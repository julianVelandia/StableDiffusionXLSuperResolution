```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "stabilityai/stable-diffusion-xl-base-1.0"
pipeline = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
).to("cuda")
```

```python
from diffusers import StableDiffusionXLPipeline


def initialize_model(model_id:str):
    """
    Inicializa el modelo
    Returns:
        pipeline: Modelo cargado y configurado.
    """
    pipeline = StableDiffusionXLPipeline.from_pretrained(
        model_id,
        torch_dtype=torch.float16,
    ).to("cuda")
    pipeline.enable_attention_slicing()

    return pipeline
```

```python
from PIL import Image
from RealESRGAN import RealESRGAN

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def super_resolve_image_realesrgan(input_path, output_path='super_resolved_image.png'):
    """
    Aplica superresolución usando RealESRGAN.

    Args:
    - input_path (str): Ruta de la imagen de entrada.
    - output_path (str): Ruta donde se guardará la imagen de salida.
    """
    model = RealESRGAN(device, scale=4)
    model.load_weights('weights/RealESRGAN_x4.pth', download=True)

    image = Image.open(input_path).convert('RGB')

    sr_image = model.predict(image)

    sr_image.save(output_path)
    print(f"Imagen superresuelta guardada en: {output_path}")
```

```python
from google.colab import files


def generate_and_save_image(pipeline, prompt, output_path="generated_image.png", super_resolution = True, width=1024, height=576):
    """
    Genera una imagen a partir de un texto.
    Args:
        pipeline: El modelo cargado.
        prompt (str): Texto descriptivo para la generación.
        output_path (str): Ruta para guardar la imagen generada.
        width (int): Ancho de la imagen generada.
        height (int): Altura de la imagen generada.
    Returns:
        None: Guarda la imagen en la ruta especificada.
    """
    with torch.no_grad():
        image = pipeline(
            prompt=prompt,
            negative_prompt="blurry, low quality, distorted, poorly rendered, cartoonish",
            width=width,
            height=height,
            guidance_scale=7.5,
            num_inference_steps=100
        ).images[0]

    image.save(output_path)
    if super_resolution:
        sr_output_path = "super_resolved_" + output_path
        super_resolve_image_realesrgan(output_path, sr_output_path)
        print("Superresolución aplicada.")
        files.download(sr_output_path)
    else:
        files.download(output_path)
```

```python
import re

def normalize_prompt(prompt, chars = 30):
    """
    Normaliza el prompt eliminando caracteres no alfanuméricos y truncándolo a un tamaño manejable.
    Args:
        prompt (str): El texto del prompt.
    Returns:
        str: Una versión normalizada y truncada del prompt.
    """
    prompt = re.sub(r"[^a-zA-Z0-9]", "_", prompt)
    return prompt[:chars]
```

```python
torch.cuda.empty_cache()
torch.cuda.memory_summary(device=None, abbreviated=False)

pipeline = initialize_model(model_id)
```

```python
import random


prompts = [
    "A vibrant carnival at night, with colorful lights illuminating Ferris wheels, carousels, and food stalls, children laughing, and a fireworks display lighting up the starry sky, ultra-realistic",
    "An alien jungle filled with towering, luminescent plants, strange creatures roaming among the foliage, and a glowing waterfall cascading into a turquoise pool, surreal and otherworldly, ultra-realistic"
]

for i, prompt in enumerate(prompts):
    normalized_prompt = normalize_prompt(prompt)
    random_number = random.randint(1000, 9999)
    output_path = f"image_{i+1}_{normalized_prompt}_{random_number}.png"
    generate_and_save_image(pipeline, prompt, output_path)

```
